```python
from __future__ import division,print_function

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

import glob
import re

files = sorted(glob.glob('energies*.dat'))

plt.figure(figsize=[14.,10.])
plt.title('')
plt.xlabel(r'$\alpha$',fontsize=15)
plt.ylabel(r'$E$',fontsize=15)
plt.xlim([0.,100.])
plt.ylim([-1.105,-1.085])
plt.grid(True)

for f in files:
    data = np.loadtxt(f)
    
    N = int(float(re.findall(r'[-+]?\d+[\.]?\d*[eE]?[-+]?\d*',f)[0]))
    Emin = np.min(data[:,1])
    if (N<9):
        print('N =',N,'  $min\left(E_{tot}(\alpha)\right)$'+': {:.5f} [Hartree]'.format(Emin),'    min(E_{tot})-$${:2f}[eV]'.format((Emin+1)*27.2114),'   alpha:',data[:,0][np.where(data[:,1] == Emin)][0])
        plt.plot(data[:,0],data[:,1],marker='o',label='N={}'.format(N))
    if (N==9):
        plt.scatter(data[:,0],data[:,1],marker='+',label='N={}'.format(N),color='orange')
plt.legend(loc='best')
plt.show()
plt.close('all')
```
